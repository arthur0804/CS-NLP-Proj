```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
source = requests.get("https://listingcenter.nasdaq.com/IssuersPendingSuspensionDelisting.aspx").text
soup = BeautifulSoup(source,'lxml')
```

```python
table = soup.find("body").find("form").find_all("div", attrs={'id': None})[3].find("div", attrs={'class': "masterwrapper"}).find("div", attrs={'id': "maincontainer"}).find("div", attrs={'id':"ctl00_tblMaster"}).find("div").find("table").find("tbody")
```

```python
# initialize a data frame
label = ['IssuerName', 'Symbol', 'Reason', 'Status', 'EffectiveDate', 'From25Date']
df = pd.DataFrame(columns = label)
```

```python
df
```

```python
# "rgRow" class
records = []
for row in table.find_all("tr", attrs = {"class" : "rgRow"}):
    recordlist = []
    for cell in row.find_all("td"):
        recordlist.append(cell.text)
    recordlist = [element.replace('\xa0', 'None') for element in recordlist]
    records.append(recordlist)
    print(recordlist)
```

```python
# add into data frame
for record in records:
    df.loc[len(df)] = record
```
